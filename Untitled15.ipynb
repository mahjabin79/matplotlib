```python
import matplotlib.pyplot as plt
import seaborn as sns


X = [2, 4, 6, 8, 10, 12, 14]
Y = [50, 100, 150, 200, 250, 300, 350]

plt.scatter(X, Y, label='Data Points')

plt.scatter(X, Y, color='green', marker='o', s=50, label='Small Green Dots')

plt.scatter(X, Y, color='red', marker='o', s=100, label='Large Red Dots')

plt.show()
```
